# Q1. What is the difference between Ordinal Encoding and Label Encoding? Provide an example of when you might choose one over the other.

# ANS:- The differnce between the __Ordinal_Encoding__ and __Label_Encoding__ lies in the way the categorical data are converted into numeric data. Talking about the ordinal Encoding , in this case , the ranks are assigned to each data based on ranking , for examples , let say we have ordinal data which has good better best . so we can easily say by our common sense that best has more ranking than better and better has more ranking than good.

# But talking about label Encoder , it assigns order to the categorical data based on their alphabets . if we have a dataset which contains rich , poor , millionare . then in this case , the rank will be assigned based on their first letters

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder , LabelEncoder
```

# Q2. Explain how Target Guided Ordinal Encoding works and provide an example of when you might use it in a machine learning project.

#ANS:- The Target Guided encoding is applied when we have to encode the target featurs which has relation some column containing categorical data . or it has some relation. let's understand with the help of example

```python
df=sns.load_dataset("tips")
```

# Q3. Define covariance and explain why it is important in statistical analysis. How is covariance calculated?

# ANS:-


#Covariance is a statistical measure that describes the extent to which two random variables change together. In other words, it measures the degree of joint variability between two variables. If the covariance between two variables is positive, it indicates that they tend to increase or decrease together. If the covariance is negative, it means that as one variable increases, the other tends to decrease.

# Covariance is important in statistical analysis for several reasons:

#Relationship Assessment:
 #Covariance helps in understanding the direction of the relationship between two variables. A positive covariance suggests a positive relationship, while a negative covariance suggests a negative relationship.

#Scaling:
#Covariance is not standardized and depends on the units of the variables involved. Therefore, it is sometimes difficult to interpret the magnitude of the covariance. To address this, correlation (which is derived from covariance) is often used, as it is a standardized measure with values between -1 and 1.

#Portfolio Analysis:
# In finance, covariance is crucial in portfolio analysis. It helps investors understand how different assets in a portfolio move in relation to each other. Diversification, a key concept in portfolio management, relies on understanding the covariance between assets to mitigate risk.

#Linear Regression:
#Covariance is used in simple linear regression analysis, where it helps determine the slope of the regression line.

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset("tips")
```

```python
np.cov(df['tip'],y=df['size'])
```
