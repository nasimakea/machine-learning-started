# Q3. Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.

# ANS:-

```python
import pandas as pd
import random

# Generate random data for 50 participants
random.seed(42)  # Setting seed for reproducibility

data = {
    'Participant_ID': [f'Person{i}' for i in range(1, 51)],
    'Hours_of_Excercise_Week': [random.uniform(1, 10) for J in range(50)],  # Random hours of exercise
    'BMI': [random.uniform(18.5, 30) for J in range(50)]  # Random BMI values
}

# Creating DataFrame
exercise_bmi_df = pd.DataFrame(data)

# Displaying the DataFrame
exercise_bmi_df.head(5)
```

```python
exercise_bmi_df.corr("pearson")
```

# Q6. A company is interested in examining the relationship between the number of sales calls made per day and the number of sales made per week. The company collected data on both variables from a sample of 30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.

# ANS:-

```python
import pandas as pd
import random

# Generate random data for 30 individuals
random.seed(42)  # Setting seed for reproducibility

data = {
    'Person_ID': [f'Person{i}' for i in range(1, 31)],
    'Sales_Calls_Per_Day': [random.randint(10, 30) for _ in range(30)],  # Random sales calls per day
    'Sales_Per_Week': [random.randint(5, 20) for _ in range(30)]  # Random sales per week
}

# Creating DataFrame
sales_data_df = pd.DataFrame(data)

# Displaying the DataFrame
print(sales_data_df.head(2))
```

```python
sales_data_df.corr("pearson")
```

```python
sales_data_df.corr("spearman")
```

```python
sales_data_df.cov()
```
